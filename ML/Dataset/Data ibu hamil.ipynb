```python
import pandas as pd
from sklearn.metrics import accuracy_score,confusion_matrix,precision_score,f1_score,recall_score 

df = pd.read_csv("/content/drive/MyDrive/Latihan Keamanan Jaringan/Data Ibu Hamil (Cleaning).csv")
```

```python
df = df[['Umur',	'Sistole',	'Diastole',	'TB',	'BB',	'IMT', 'Lingkar Perut', 'Diagnosis']]
```

```python
df['Umur'] = df['Umur'].str.replace(r'\D', '')
df
```

```python
df.columns
```

```python
df['Diagnosis'].value_counts()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
X = df.apply(le.fit_transform)
X
```

```python
x = X[['Umur',	'Sistole',	'Diastole',	'TB',	'BB',	'IMT',	'Lingkar Perut']]
```

```python
y = df[['Diagnosis']].values     
```

```python
 x
```

```python
y[0:5]
```

```python
from sklearn.model_selection import train_test_split
x1_train, x1_test, y_train, y_test = train_test_split( x, y, test_size=0.2, random_state=0)
print ('Train:', x1_train.shape, y_train.shape)
print ('Test:', x1_test.shape, y_test.shape)
```

```python
from sklearn.svm import SVC
# from sklearn.naive_bayes import GaussianNB
model = SVC(kernel='rbf', gamma='auto')
# model = GaussianNB() 
model.fit(x1_train, y_train)
Hasil_Prediksi=model.predict(x1_test)
```

```python
print("Akurasinya :",(accuracy_score(Hasil_Prediksi,y_test))*100, "%")
```

```python
print(model.predict([[214,	62,	41,	27,	117,	177,	20]])[0])
```

```python
import pandas as pd

import warnings
warnings.filterwarnings('ignore')
```

```python
df1 = pd.read_csv('Maternal Health Risk Data Set.csv')
```

```python
df1
```

```python
df2 = pd.read_csv('Data Ibu Hamil (Cleaning).csv')
```

```python
df2 = df2[['Umur', 'Sistole', 'Diastole', 'Diagnosis']]
```

```python
# yang tidak ada BS	BodyTemp	HeartRate

df2['BS'] = df1['BS'].mean()
df2['BodyTemp'] = df1['BodyTemp'].mean()
df2['HeartRate'] = df1['HeartRate'].mean()
```

```python
df2
```

```python
# ambil umur nya aja

df2['Umur'] = df2['Umur'].str.split(pat ='Tahun', expand = True)[0]
```

```python
df2
```

```python
df1.columns
```

```python
df1 = df1.replace('high risk', 0)
df1 = df1.replace('mid risk', 0)
df1 = df1.replace('low risk', 1)

df2 = df2.replace('Z34', 1)
df2 = df2.replace('Z39', 0)
```

```python
df2 = df2.rename(columns = {'Umur':'Age',
                       'Sistole':'SystolicBP',
                       'Diastole':'DiastolicBP',
                       'BS':'BS',
                       'BodyTemp':'BodyTemp',
                       'HeartRate':'HeartRate',
                       'Diagnosis':'RiskLevel'})
```

```python
df1.head(1)
```

```python
df2.head(1)
```

```python
df_final = pd.concat([df1,df2])
```
